```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# crete an empty canvas

height, width = 500, 700

# selecting canvas colour as blue acc to the bgr formate
blue = (255, 127, 0)

canvas = np.full((height, width, 3), blue, dtype=np.uint8)

plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
# plt.imshow(canvas)
```

Draw a line

```python
cv2.line(canvas, (50,50), (650, 450), (255,255,255), 10)
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```

Draw a rectangle

```python
cv2.rectangle(canvas, (500,100), (300,300), (0,255,255), -18)
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```

Draw a circle


```python
cv2.circle(canvas, (500,150), 100, (0,0,255), 5)
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```

Draw an ellipse 

```python
cv2.ellipse(canvas, (350,150), (100,50), 45, 0, 360, (0,255,0), 4)
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```

Draw polygon


```python
#Xn, Yn, where n>2

points = np.array([[150,400], [300,400], [255, 300]], np.int32)

new_reshaped_point = points.reshape((-1,1,2))
print(new_reshaped_point)

cv2.polylines(canvas, [new_reshaped_point],True, (155,0,255), 3)
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```
